```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/canada_per_capita_income.csv")
```

```python
df
```

Simple scatter plot of the data

```python
plt.scatter(df['year'], df['per capita income (US$)'], color= 'green', marker='*')
plt.xlabel('year')
plt.ylabel('income')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])#so what will be the value in x year, reg.predict x!
```

```python
reg.predict([[2000]])
```

#y=mx+b, m=coeff,b=intercept

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=mx+b- proof of linear regression calculation
y=828.46507522*2000-1632210.7578554575
y
```

#representation of the plot

```python
plt.scatter(df['year'], df['per capita income (US$)'], color= 'green', marker='*')
plt.xlabel('year')
plt.ylabel('income')
plt.plot(df['year'], reg.predict(df[['year']]), color='blue')
```
